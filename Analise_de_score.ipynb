# Introduo ao Score de Crditos 

O Score de Crditos  uma ferramenta essencial para entender sua sade financeira. Ele representa uma pontuao numrica que reflete o risco de crdito de um indivduo, que significa as chances de um indivduo que tomou crdito do banco vir a se tornar um inadimplente. Neste notebook de Python, vamos explorar o que  o Score de Crditos, como  calculado e qual sua importncia para as instituies financeiras e tomadores de crdito.

 Como  calculado? O Score de Crditos  calculado com base em diversas informaes financeiras, como histrico de pagamentos, dvidas em aberto, prazos de pagamento, entre outros fatores. Quanto mais alto for o score, maior a confiana das instituies em sua capacidade de pagamento.

 Importncia do Score de Crditos: Um bom score  vital para solicitar emprstimos, financiamentos ou at mesmo para obter melhores condies em compras a prazo. Ele pode influenciar nas taxas de juros oferecidas pelas instituies financeiras, impactando diretamente em suas decises financeiras.

Um bom Score  resultado de:

*   Contas pagas em dia 
*   Cadastro atualizado 
*   Sem dvidas em atrado 
*   Uso do crdito de forma consciente
